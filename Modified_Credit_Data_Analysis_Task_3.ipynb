# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()
dataset['Gender'] = label.fit_transform(dataset['Gender'])
dataset['Student'] = label.fit_transform(dataset['Student'])
dataset['Married'] = label.fit_transform(dataset['Married'])
dataset['Ethnicity'] = label.fit_transform(dataset['Ethnicity'])

dataset
```

# Model Training

```python
from sklearn.model_selection import train_test_split

array = dataset.values
x = array[:,:-1]
y = array[:,-1]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state=42)
```

# Training and Testing the Model using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report, accuracy_score,confusion_matrix

accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

So, the accuracy obtained using Logistic Regression is around 95.8%

# Training and Testing the Model using Support Vector Machine

```python
from sklearn.svm import SVC

model = SVC()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

So, the accuracy obtained using Support Vector Machine is approximately 89%

Therefore, Logistic Regression worked better than Support Vector Machine for the dataset
